```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame({"a" : [4 ,5, 6],"b" : [7, 8, 9],"c" : [10, 11, 12]},index = [1, 2, 3])
df1
```

### Specify values for each row:

```python
df2 = pd.DataFrame([[4, 7, 10],[5, 8, 11],[6, 9, 12]],index=[1, 2, 3],columns=['a', 'b', 'c'])
df2
```

### Create DataFrame with a MultiIndex:

```python
df = pd.DataFrame({"a" : [4 ,5, 6],"b" : [7, 8, 9],"c" : [10, 11, 12]},index = pd.MultiIndex.from_tuples([('d',1),('d',2),('e',2)],
names=['n','v']))
df
```

### Reshaping Data: Change the layout of a data set:

```python
#Gather columns into rows.
pd.melt(df)
```

```python
#Append columns of DataFrames
df3=pd.concat([df1,df2], axis=1)
df3
```

```python
#Rename the columns of a DataFrame
df3.rename(columns = {'c':'c_year'})
```

```python
#Sort the index of a DataFrame
df3.sort_index()
```

```python
#Reset index of DataFrame to row numbers, moving index to columns.
df3.reset_index()
```

```python
host_names = ['Maarten','Fronk & Lieve','Elke','Francis','Kristof',
              'Liesbet','Chloé','Klara','Maarten','Kristina']
 
listing_count = [1, 7, 1, 2, 1, 1, 1, 1, 1, 1]
review_count = [25, np.nan, 4, 59, 3, 89, 3, 2, 7, np.nan]
accommodates = [2, 2, 3, 2, 2, 4, 2, 5, 4, 2]
 
index=['2014-04-01', '2012-02-27', '2013-05-11', '2012-04-17', 
                               '2012-10-13', '2011-01-29', '2014-03-15', '2014-05-22', 
                               '2012-07-24', '2014-02-17']
 
data = {'host_name': host_names, 'listing_count': listing_count, 
        'review_count': review_count,'accommodates': accommodates}
host_df = pd.DataFrame(data, index=index)
host_df
```

```python
# Default behavior 
# NaN values will be in last
# In any sorting order
host_df.sort_values(by=['review_count'], na_position='first')
```

```python

# Sorting
# Level 0 (Host Name) Descending
# Level 1 (Year of joining) Ascending
host_df.sort_index(level=[0, 1], ascending=[False, True])
```
