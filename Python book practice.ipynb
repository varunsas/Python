### More recursion

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        recurse = factorial(n-1)
        result = n*recurse
        return result
```

```python
factorial(6)
```

```python
def fibonacci(n):
    if n == 0 or n == 1:
        return 1
    else:
        return fibonacci(n-1)+ fibonacci(n-2)
```

```python
fibonacci(7)
```

```python
def factorial(n):
    if not isinstance(n,int):
        print("Factorial is only defined for integers.")
        return -1
    elif n < 0:
        print("Factorial is only defined for positive integers.")
        return -1
    elif n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(4)
```

```python
factorial("fred")
```

```python
factorial(-2)
```
