## Creating a pie chart  using python 3 graphical libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv("C:\\Users\\HP\\Downloads\\Ebola.csv",encoding= 'latin-1')
```

## Creating a stacked area plot using Python seaborn library 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df2=pd.read_csv("C:\\Users\\HP\\Downloads\\WashingtonDCCrime.csv",encoding='latin-1')
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
summary=pd.crosstab(df2['Year'],df2['Crime'],values=df2['Rate'],aggfunc=np.sum)
```

```python
summary.head()
```

```python
plt.rcParams["figure.figsize"]=[10,8]
sns.set_context("notebook",font_scale=1.5,rc={"font.size":14,"axes.titlesize":16,"axes.labelsize":12})
summary.plot(kind="area",stacked=True)
plt.title("Stacked area plot of washington DC violent crime data")
plt.ylabel("Crime Rate per 100000s")
plt.xlabel("Year reported")
plt.savefig('stacked_area.png',dpi=1200)
plt.show()
```

## Creating a stacked Bar of the missing migrants data using Python seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df10= pd.read_csv("C:\\Users\\HP\\Downloads\\Migrants.csv",encoding= 'latin-1')
```

```python
df10.head()
```

### Subset the migrants death to certain region

```python
df11= df10.loc[df10['Region_of_Incident'].isin(['Horn of Africa','Mediterranean','Middle East','North Africa','Sub-Saharan Africa'])]
```

### Create a crosstab tables with the total deaths and missings

```python
summary=pd.crosstab(df11['Region_of_Incident'],df11['Reported_Year'],values=df11['Total_Dead_and_Missing'],aggfunc=np.sum)
```

```python
summary.head()
```

### Use Seaborn to style and pandas to plot the bar graphs

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
plt.rcParams["figure.figsize"] = [10,6]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16})
sns.set_style("darkgrid")
summary.plot(kind="bar", stacked=False,color=flatui)
plt.title('Number of deaths or missing African migrants \n Period of 2014 thru 2017')
plt.ylabel('Number of deaths or missing migrants')
plt.xlabel('Region of casualties')
plt.xticks(rotation=-45)
plt.show()
```

```python
flatui = ["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"]
plt.rcParams["figure.figsize"] = [10,6]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":16,"axes.titlesize":16,"axes.labelsize":16})
sns.set_style("darkgrid")
summary.plot(kind="bar", stacked=True,color=flatui)
plt.title('Number of deaths or missing African migrants \n Period of 2014 thru 2017')
plt.ylabel('Number of deaths or missing migrants')
plt.xlabel('Region of casualties')
plt.xticks(rotation=-45)
plt.show()
```
